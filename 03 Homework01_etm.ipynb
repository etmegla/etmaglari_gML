## 4. Importing the OBJs

```python
OBJ_PATH = r"C:\Users\etmaglari\IAAC\etmaglari_gML\Rhino_Geometry _etm.obj"

objects  = Topology.ByOBJPath(OBJ_PATH, selfMerge=False)
topology = objects[0] if isinstance(objects, list) else objects
print(f"Loaded: {type(topology).__name__}")

clean = Topology.RemoveCoplanarFaces(
    topology,
    epsilon=0.1,
    tolerance=0.001
)
print(f"Faces before: {len(Topology.Faces(topology))}")
print(f"Faces after RemoveCoplanarFaces: {len(Topology.Faces(clean))}")
```

## 6. Build CellComplex from cleaned geometry

```python
faces = Topology.Faces(clean)
cc = CellComplex.ByFaces(faces, tolerance=0.001)
cells = Topology.Cells(cc)
print(f"CellComplex: {len(cells)} cells, {len(Topology.Faces(cc))} faces, {len(Topology.Edges(cc))} edges")
```

## 7. Primal Graph

Vertices at the geometric corners of the CellComplex; edges along wall boundaries.

```python
primal_verts = Topology.Vertices(cc)
primal_edges = Topology.Edges(cc)
print(f"Primal — Vertices: {len(primal_verts)}, Edges: {len(primal_edges)}")

Topology.Show(
    cc,
    faceColor=[210, 210, 250], faceOpacity=0.15,
    edgeColor="red", edgeWidth=2,
    vertexColor="red", vertexSize=6,
    showVertices=True,
    backgroundColor="black",
    width=800, height=600,
    renderer=renderer
)
```

## 8. Dual Graph

One vertex per room (cell centroid); an edge between every pair of rooms that share a wall face.

```python
g_dual = Graph.ByTopology(cc)
print(f"Dual — Nodes: {len(Graph.Vertices(g_dual))}, Edges: {len(Graph.Edges(g_dual))}")

for v in Graph.Vertices(g_dual):
    Topology.SetDictionary(v, Dictionary.ByKeysValues(["size", "color"], [18, "red"]))
for e in Graph.Edges(g_dual):
    Topology.SetDictionary(e, Dictionary.ByKeysValues(["width", "color"], [3, "white"]))

Topology.Show(
    cc, g_dual,
    faceColor="white", faceOpacity=0.05,
    edgeColor="white", edgeWidth=0.5,
    vertexSizeKey="size", vertexColorKey="color",
    edgeWidthKey="width", edgeColorKey="color",
    showVertices=True,
    backgroundColor="black",
    width=800, height=600,
    renderer=renderer
)
```

## 11. Add Apertures to CellComplex

```python
all_apertures = windows + doors
cc_apts = Topology.AddApertures(cc, all_apertures)

# Assign a palette colour + id to every cell for downstream graph colouring
all_cells = Topology.Cells(cc_apts)
palette   = ["#E63946","#F4A261","#2A9D8F","#457B9D","#A8DADC",
             "#264653","#8338EC","#3A86FF","#FB5607","#FFBE0B"]
for i, cell in enumerate(all_cells):
    color = palette[i % len(palette)]
    Topology.SetDictionary(cell, Dictionary.ByKeysValues(["color", "id"], [color, i]))

print(f"CellComplex: {len(all_cells)} rooms, {len(all_apertures)} apertures")

Topology.Show(
    cc_apts, win_cluster, door_cluster,
    faceColor=[210, 210, 250], faceOpacity=0.08,
    edgeColor="white", edgeWidth=0.5,
    showVertices=False,
    backgroundColor="black",
    width=800, height=600,
    renderer=renderer
)
```

## 12. Access + Adjacency Graph

Rooms connected when they **share a wall** (direct=True), share an **interior aperture** (iaSharedApertures=True), or have an **exterior aperture** (	oExteriorApertures=True). Node size scales with room surface area.

```python
m = Graph.ByTopology(
    cc_apts,
    direct=True,
    viaSharedApertures=True,
    toExteriorApertures=True
)
print(f"Graph vertices: {len(Graph.Vertices(m))}")
print(f"Edges:          {len(Graph.Edges(m))}")

areas    = [Cell.SurfaceArea(cell) for cell in all_cells]
min_area = min(areas)
max_area = max(areas)

for v in Graph.Vertices(m):
    vx, vy, vz = Vertex.X(v), Vertex.Y(v), Vertex.Z(v)
    best_color = "gray"
    best_size  = 8
    best_dist  = float("inf")
    for cell, area in zip(all_cells, areas):
        c    = Topology.Centroid(cell)
        dist = ((vx - Vertex.X(c))**2 + (vy - Vertex.Y(c))**2 + (vz - Vertex.Z(c))**2)**0.5
        if dist < best_dist:
            best_dist  = dist
            cd         = Topology.Dictionary(cell)
            best_color = Dictionary.ValueAtKey(cd, "color") if cd else "gray"
            if best_color == "gray":
                best_size = 8
            else:
                best_size = 12 + int(48 * (area - min_area) / (max_area - min_area + 1))
    Topology.SetDictionary(v, Dictionary.ByKeysValues(["size", "color"], [best_size, best_color]))

for e in Graph.Edges(m):
    Topology.SetDictionary(e, Dictionary.ByKeysValues(["width", "color"], [2, "black"]))

Topology.Show(
    [cc_apts, m, win_cluster, door_cluster],
    vertexSizeKey="size",
    vertexColorKey="color",
    edgeWidthKey="width",
    edgeColorKey="color",
    faceOpacity=0.15,
    backgroundColor="white",
    width=1000, height=1000,
    renderer=renderer
)
```

## 13. Room Centroid to Window and Door Graph

```python
from collections import defaultdict

cells_cc = Topology.Cells(cc)

def dist3(a, b):
    return ((a[0]-b[0])**2 + (a[1]-b[1])**2 + (a[2]-b[2])**2)**0.5

# Pre-compute shared (interior) face centroid keys once
shared_keys = set()
for f in Topology.Faces(cc):
    adjacent = Topology.SuperTopologies(f, hostTopology=cc, topologyType="cell")
    if adjacent and len(adjacent) > 1:
        fc = Topology.Centroid(f)
        shared_keys.add((round(fc.X(), 2), round(fc.Y(), 2), round(fc.Z(), 2)))
print(f"Shared (interior) faces: {len(shared_keys)}")

def nearest_room(apt_face):
    """Return index of room whose nearest exterior face is closest to apt centroid."""
    ac = Topology.Centroid(apt_face)
    ap = (ac.X(), ac.Y(), ac.Z())
    best_d, best_i = float("inf"), 0
    for i, cell in enumerate(cells_cc):
        for f in Topology.Faces(cell):
            fc  = Topology.Centroid(f)
            key = (round(fc.X(), 2), round(fc.Y(), 2), round(fc.Z(), 2))
            if key in shared_keys:
                continue
            d = dist3(ap, (fc.X(), fc.Y(), fc.Z()))
            if d < best_d:
                best_d, best_i = d, i
    return best_i

# ── Room centroid nodes: large, palette colour ───────────────────────────────
room_verts = []
for i, cell in enumerate(cells_cc):
    rc    = Topology.Centroid(cell)
    color = palette[i % len(palette)]
    v     = Vertex.ByCoordinates(rc.X(), rc.Y(), rc.Z())
    Topology.SetDictionary(v, Dictionary.ByKeysValues(["size", "color"], [18, color]))
    room_verts.append(v)

# ── Window centroid nodes: small cyan ────────────────────────────────────────
win_verts = []
for w in windows:
    ac = Topology.Centroid(w)
    v  = Vertex.ByCoordinates(ac.X(), ac.Y(), ac.Z())
    Topology.SetDictionary(v, Dictionary.ByKeysValues(["size", "color"], [7, "#4CC9F0"]))
    win_verts.append(v)

# ── Door centroid nodes: small gold ──────────────────────────────────────────
door_verts = []
for d_face in doors:
    ac = Topology.Centroid(d_face)
    v  = Vertex.ByCoordinates(ac.X(), ac.Y(), ac.Z())
    Topology.SetDictionary(v, Dictionary.ByKeysValues(["size", "color"], [7, "#FFD700"]))
    door_verts.append(v)

# ── Assign apertures to rooms ─────────────────────────────────────────────────
room_to_wins  = defaultdict(list)
room_to_doors = defaultdict(list)
for j, w in enumerate(windows):
    room_to_wins[nearest_room(w)].append(j)
for j, d_face in enumerate(doors):
    room_to_doors[nearest_room(d_face)].append(j)

# ── Build edges (dicts stay on the original Edge objects) ────────────────────
edges_rw = []
for ri, win_idxs in room_to_wins.items():
    rv = room_verts[ri]
    for j in win_idxs:
        e = Edge.ByVertices([rv, win_verts[j]])
        Topology.SetDictionary(e, Dictionary.ByKeysValues(["width", "color"], [2, "#4CC9F0"]))
        edges_rw.append(e)
for ri, door_idxs in room_to_doors.items():
    rv = room_verts[ri]
    for j in door_idxs:
        e = Edge.ByVertices([rv, door_verts[j]])
        Topology.SetDictionary(e, Dictionary.ByKeysValues(["width", "color"], [2, "#FFD700"]))
        edges_rw.append(e)

print("Windows per room:")
for ri, idxs in sorted(room_to_wins.items()):
    print(f"  Room {ri}: {len(idxs)} window(s)")
print("Doors per room:")
for ri, idxs in sorted(room_to_doors.items()):
    print(f"  Room {ri}: {len(idxs)} door(s)")
print(f"Edges built: {len(edges_rw)}")
```

```python
# Bipartite graph: room centroids -> window centroids (cyan)
#                  room centroids -> door centroids   (gold)
bipartite = room_verts + win_verts + door_verts + edges_rw

Topology.Show(
    bipartite + [cc],
    faceColor="white", faceOpacity=0.05,
    edgeColor="white", edgeWidth=0.5,
    vertexColor="black",
    vertexSizeKey="size", vertexColorKey="color",
    edgeWidthKey="width", edgeColorKey="color",
    showVertices=True,
    backgroundColor="black",
    width=900, height=700,
    renderer=renderer
)
```
