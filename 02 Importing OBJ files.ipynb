## 4. Import the OBJ file

```python
from pathlib import Path

candidate_paths = [
    Path.cwd() / "Rhino_Geometry _etm.obj",
    Path.cwd().parent / "Rhino_Geometry _etm.obj",
    Path("c:/Users/etmaglari/IAAC/etmaglari_gML/Rhino_Geometry _etm.obj"),
]
obj_path = next((path for path in candidate_paths if path.exists()), None)
if obj_path is None:
    raise FileNotFoundError("Could not find Rhino_Geometry _etm.obj in the workspace.")

objects = Topology.ByOBJPath(str(obj_path))
print(f"Loaded OBJ from: {obj_path}")
print(f"Imported {len(objects)} object(s)")
```

## 6. Show the geometry

```python
Topology.Show(objects,
              faceColor=[210,210,250],
              faceOpacity=0.2,
              edgeColor="white",
              edgeWidth=3,
              showVertices=False,
              backgroundColor="black",
              width=800,
              height=600,
              renderer = renderer)
```
