## 4. Import the OBJ file

```python
from pathlib import Path

workspace_root = Path.cwd()
candidate_names = [
    "Rhino_Geometry_etm.obj",
    "Rhino_Geometry _etm.obj",
    "Rhino_Geometry.obj",
]

candidate_paths = []
for folder in [workspace_root, workspace_root.parent, Path("c:/Users/etmaglari/IAAC/etmaglari_gML")]:
    for name in candidate_names:
        candidate_paths.append(folder / name)

obj_path = next((path for path in candidate_paths if path.exists()), None)
if obj_path is None:
    searched_paths = "\n".join(str(path) for path in candidate_paths)
    raise FileNotFoundError(
        "Could not find an OBJ file. Searched:\n" + searched_paths
    )

objects = Topology.ByOBJPath(str(obj_path))
print(f"Loaded OBJ from: {obj_path}")
print(f"Imported {len(objects)} object(s)")
```

## 6. Show the geometry

```python
Topology.Show(objects,
              faceColor=[210,210,250],
              faceOpacity=1,
              edgeColor="white",
              edgeWidth=3,
              showVertices=False,
              backgroundColor="black",
              width=800,
              height=600,
              renderer = renderer)
```
